```python
# You can enter max 20 numbers one by one and end the loop to enter the number with 0.
# If you enter a string then it warns you and assure to enter a number.


def get_num ():             # the function to enter the numbers.It appends the numbers to the list one by one.
    count =1
    numbers_list = []
    print('Please pay attention..!')
    print('You can enter max 20 numbers. If you want to end the loop, please enter 0 ..!')
    
    while (count<21):
        number = input('Please enter a number :')
        if number.isdecimal()==False:
            while number.isdecimal()==False:
                print('Please enter only numbers..!')
                number = input('Please enter a number :')        
        
        number= int(number)
        if number==0:
            print('You entered %d numbers..!' %(count-1))
            break
        if number ==1:
            print ('1 is not prime number..!')
            continue
            
        numbers_list.append(number)
        if count==19:
            print('You will enter the last number...!')
        count +=1
    return numbers_list


def prime(number):           # The function controls the numbers in the list one by one wheather it is a prime number.
    dividend =[2,3,5,7]
    if number not in dividend:
        dividend.append(number)
    count=0
    
    for x in dividend:
        if (number%x == 0):
            count +=1 
    if count==1:
        print(number, ' is prime number.')    # If it is a prime number, then it prints the prime number.
        return (number, True)
    else:
        return (number, False)

    
def num_list(data):    # From the given numbers it prints True if the number is prime, otherwise False.
    my_list = []
    for x in data:
        say,deger = prime(x)
        my_list.append([say,deger])
    [print(x[0],x[1]) for x in my_list] 


nums= get_num()
num_list(nums)
```
